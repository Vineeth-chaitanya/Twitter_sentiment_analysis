Converting target value 4 to 1

```python
twitter_data.replace({'target':{4:1}}, inplace=True)
```

**Stemming**
This is a process to reduce a word to its root word.(ex: actor,actress-> act)


```python
port_stem= PorterStemmer()
```

defining a function

```python
def stemming(content):

  stemmed_content= re.sub('[^a-zA-Z]',' ', content) #removing all the special char in text column
  stemmed_content= stemmed_content.lower() #converting all of them into lowecase
  stemmed_content= stemmed_content.split() #taking each word seperately
  stemmed_content= [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')] #if the word is not from stopwords it is joined
  stemmed_content= ' '.join(stemmed_content)

  return stemmed_content
```

```python
twitter_data['stemmed_content'] = twitter_data['text'].apply(stemming)
```

```python
twitter_data.head()
```

```python
print(twitter_data['stemmed_content'])
```

```python
print(twitter_data['target'])
```

```python
#seperating the data and the values
x=twitter_data['stemmed_content'].values
y=twitter_data['target'].values
```

```python
print(x)
```

```python
print(y)
```

```python
#splitting the data into training and testing
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
#converting textual data to numerical data

vectorizer= TfidfVectorizer()

x_train= vectorizer.fit_transform(x_train)
x_test= vectorizer.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

**Training the machine learning model

Logistic Regression**

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

**Model Evaluation

Accuracy score**

```python
#accuracy score on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
```

```python
print('The training data accuracy is :',training_data_accuracy)
```

```python
#accuracy score on test data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(y_test,x_test_prediction)
```

```python
print('The testing data accuracy is:',testing_data_accuracy)
```

**Model Accuracy is 77.8%**

when training data accuracy is more than testing data accuracy then the model is overfitting model.
